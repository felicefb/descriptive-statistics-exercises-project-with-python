# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO

import numpy as np

# Generate 100 elements from a normal distribution (mean=0, std=1)
normal_dist = np.random.normal(loc=0, scale=1, size=100)

# Generate 100 elements from a chi-square distribution with 3 degrees of freedom
chi_square_dist = np.random.chisquare(df=3, size=100)

# Function to compute key descriptive statistics
def describe_data(data, name):
    print(f"\n{name} Statistics:")
    print(f"Mean: {np.mean(data):.4f}")
    print(f"Median: {np.median(data):.4f}")
    print(f"Standard Deviation: {np.std(data, ddof=1):.4f}")
    print(f"Variance: {np.var(data, ddof=1):.4f}")
    print(f"Min: {np.min(data):.4f}")
    print(f"Max: {np.max(data):.4f}")
    print(f"25th Percentile: {np.percentile(data, 25):.4f}")
    print(f"50th Percentile (Median): {np.percentile(data, 50):.4f}")
    print(f"75th Percentile: {np.percentile(data, 75):.4f}")

# Display statistics for both distributions
describe_data(normal_dist, "Normal Distribution")
describe_data(chi_square_dist, "Chi-Square Distribution")

import numpy as np
import matplotlib.pyplot as plt

# Generate 100 elements from a normal distribution (mean=0, std=1)
normal_dist = np.random.normal(loc=0, scale=1, size=100)

# Generate 100 elements from a chi-square distribution with 3 degrees of freedom
chi_square_dist = np.random.chisquare(df=3, size=100)

# Create subplots: one for each distribution
plt.figure(figsize=(12, 5))

# Histogram for Normal Distribution
plt.subplot(1, 2, 1)
plt.hist(normal_dist, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)

# Histogram for Chi-Square Distribution
plt.subplot(1, 2, 2)
plt.hist(chi_square_dist, bins=15, color='salmon', edgecolor='black', alpha=0.7)
plt.title("Chi-Square Distribution (df=3)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)

# Show the plots
plt.tight_layout()
plt.show()

````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO

import math

data = [4, 2, 5, 8, 6]

# Step 1: Calculate the mean
mean_value = sum(data)  /len(data)

# Step 2: Compute the squared differences
squared_diffs = [(x - mean_value) ** 2 for x in data]

# Step 3: Compute the variance
variance = sum(squared_diffs) / len(data)

# Step 4: Compute the standard deviation
std_dev = math.sqrt(variance)

print("Standard Deviation:", round(std_dev, 4))

````
